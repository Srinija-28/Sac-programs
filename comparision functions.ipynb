```python
#9.COMPARISON FUNCTIONS
import numpy as np
a1=np.array([15,21,31,8,5,4])
a2=np.array([11,41,70,10,3,1])
print("element wise greater than comparison",np.greater(a1,a2))
print("element wise greater than comparison",np.greater(a2,a1))
print("element wise less than comparison",np.less(a1,a2))
print("element wise less than comparison",np.less(a2,a1))
print("element wise equality comparison",np.equal(a1,a2))
```
