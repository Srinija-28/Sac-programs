```python
#5.LINEAR ALGEBRA FUNCTIONS
import numpy as np
a=np.array([[1,2],[6,7]])
b=np.array([[4,7],[3,7]])
print("inverse of matrix a",np.linalg.inv(a))
print("inverse of matrix b",np.linalg.inv(b))
print("determinant of matrix a",np.linalg.det(a))
print("determinant of matrix b",np.linalg.det(b))
print("rank of matrix a",np.linalg.matrix_rank(a))
print("rank of matrix a",np.linalg.matrix_rank(b))
print("matrix multiplication of a and b",np.dot(a,b))
print("eigenvalues and eigenvectors of matrix a\n",np.linalg.eig(a))
print("eigenvalues and eigenvectors of matrix b\n",np.linalg.eig(b))
print("Singular value decompostion of matrix a",np.linalg.svd(a))
print("Singular value decompostion of matrix b",np.linalg.svd(b))
```
