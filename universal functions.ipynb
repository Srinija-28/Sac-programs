```python
#8.UNIVERSAL FUNCTIONS
#→Arithematic functions
import numpy as np
arr1=np.array([[1,2,3],[6,7,8]])
arr2=np.array([[4,7,6],[3,7,1]])
print("Addition:\n",np.add(arr1,arr2))
print("Subtraction:\n",np.subtract(arr1,arr2))
print("Multiplication:\n",np.multiply(arr1,arr2))
print("Division:\n",np.divide(arr1,arr2))
#→Trigonometric functions
import numpy as np
angles_degrees=np.array([120,35,180])
angles_radians=np.radians(angles_degrees)
print("sine:\n",np.sin(angles_radians))
print("cosine:\n",np.cos(angles_radians))
print("tangent:\n",np.tan(angles_radians))
print("convert degrees to radians",np.radians(angles_degrees))
#→Exponential & logarithmic functions
import numpy as np
arr1=np.array([1,2,3,6,7])
print('exponential of elements',np.exp(arr1))
print("natural logrithm of elements",np.log(arr1))
print("log base 10 of elemnts",np.log10(arr1))
#—>Rounding functions
import numpy as np
a=np.array([1.9,2.4,3.6,4.9])
print("floor of elements",np.floor(a))
print("ceil of elements",np.ceil(a))
print("round of elements",np.round(a, decimals=1))
```
