```python
#11. FILE READING AND WRITING
#â†’Saving and loading text files
import numpy as np
#creates a numpy array
A=np.array([[1,2,3],[4,5,6]])
#save it to a text file
np.savetxt("array.txt",A,delimiter=",",fmt="%d")
print("Array saved to array.txt")
#loading an array from a text file
B=np.loadtxt("array.txt", delimiter=",",dtype=int)
print("loaded array:\n",B)
```
