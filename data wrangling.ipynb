```python
#→Hierarchical Indexing
import pandas as pd
index=pd.MultiIndex.from_tuples([('USA','NY'),('USA','CA'),('INDIA','DELHI')],names=['Country','city'])
df=pd.DataFrame({'Population':[8.6,39.5,18.9]},index=index)
print(df)
#→Combing and merging Datasets
import pandas as pd
# DataFrames
df1 = pd.DataFrame({'Emp_ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'Emp_ID': [1, 2, 4], 'Salary': [50000, 60000, 70000]})
print("Original DF1",df1)
print("Original DF2",df2)
#Inner join
result=pd.merge(df1,df2,on='Emp_ID',how='inner')
print("Inner join:\n",result)
#left join
left_result=pd.merge(df1,df2,on='Emp_ID',how='left')
print("Left join:\n",left_result)
#Right join
right_result=pd.merge(df1,df2,on='Emp_ID',how='right')
print("Right join:\n",right_result)
#Full join
full_join=pd.merge(df1,df2,on='Emp_ID',how='outer')
print("Full join:\n",full_join)
#Concatenation Combine Row-Wise(Like UNION in SQL)
import pandas as pd
df1=pd.DataFrame({
    'ID':[1,2],
    'Name':['Charlie','David']
})
df2=pd.DataFrame({
    'ID':['apple','ball'],
    'Name':['Charlie','David']
})
combined_rows=pd.concat([df1,df2],axis=0)
print("After Combining Rows:\n",combined_rows)
#Combine Column-Wise(Add new columns)
df3=pd.DataFrame({
    'Age':[25,30]
})
combined_cols=pd.concat([df1,df3],axis=1)
print("After Combining Columns:\n",combined_cols)
#Reshape Pivot
data=pd.DataFrame({
    'Month':['Jan','Jan','Feb','Feb'],
    'Product':['A','B','A','B'],
    'Sales':[100,200,150,250]
})
print("Original DataFrame\n",data)
pivoted=data.pivot(index='Month',columns='Product',values='Sales')
print("pivot elements\n",pivoted)
#Melt for reshape
import pandas as pd
wide_df=pd.DataFrame({
    'Month':['Jan','Feb'],
    'A':[100,150],
    'B':[200,250]
})
melted=pd.melt(wide_df,id_vars='Month',var_name='Product',value_name='Sales')
print("Melt:\n",melted)
#Stack and Unstack
data={
    'Math':[90,85],
    'Science':[88,92]
}
df=pd.DataFrame(data)
df.index=['Student1','Student2']
print("Original DataFrame:\n",df)
#Stack the DataFrame
stacked_df=df.stack()
print("\nStacked DatFrame:\n",stacked_df)
#Unstacked the Stacked DataFrame
unstacked_df=stacked_df.unstack()
print("\nUnstacked DataFrame:\n",unstacked_df)

```
