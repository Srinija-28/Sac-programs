```python
#15.GROUPING AND AGGREGATION
import pandas as pd
data={"Name":['Alice','Bob','Charlie','David','Eve','Alice','Bob'],'Age':[25,30,35,40,28,25,30],'Salary':[50000,60000,55000,70000,62000,51000,62000],'Allowances':[2000,2000,2000,2000,2500,2200,2100]}
df=pd.DataFrame(data)
print("Actual data frame:\n",df)
print("Groupmean:\n",df.groupby('Name').mean())
print("Group by Name=mean&sum:\n",df.groupby('Name').agg(['mean','sum']))
#create a pivot table with salary as values.grouped by name(index)
pivotTable=df.pivot_table(values='Salary',index='Name')
print("pivot table:\n",pivotTable)
```
