```python
#14.WORK WITH FOLLOWING FUNCTIONS IN SERIES AND DATA FRAME
import pandas as pd
data={"Name":['Alice','Bob','Charlie','David','Eve'],'Age':[25,30,35,40,28],'Salary':[50000,60000,55000,70000,62000],'Allowances':[2000,2000,2000,2000,2500]}
df=pd.DataFrame(data)
print("Actual data frame:\n",df)
print("Sorted salary:\n",df.sort_values('Salary',ascending=True))
print("Sorting Multiple columns:\n",df.sort_values(['Age','Salary']))
print("Setting cloumn as index:\n",df.set_index('Age'))
print("Data frame after reset of index:\n",df.reset_index())
```
