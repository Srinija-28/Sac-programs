```python
#4.MATHEMATICAL OPERATIONS
#→Element-wise Operations
import numpy as np
A=np.array([20,40,60,80])
print("elements in A:",A)
print("square of A:",np.square(A))
print("power of A:",np.power(A,2))
print("squareroot of A:",np.sqrt(A))
print("adding A+value:",A+2)
print("multiplying A*value:",A*3)
#—>Statistical & Aggregation functions
import numpy as np
arr=np.array([10,20,30,40,50])
print("sum of elements",np.sum(arr))
print("mean of elements",np.mean(arr))
print("median of elements",np.median(arr))
print("Standard deviation of elements",np.std(arr))
print("variance of elements",np.var(arr))
print("max value of elements",np.max(arr))
print("min value of elements",np.min(arr))
print("product of elements",np.prod(arr))

```
