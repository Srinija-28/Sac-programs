```python
#2.ARRAY INDEXING & SLICING
#CREATING AN ARRAY
import numpy as np
arr1D=np.array([5,11,8,20,55])
arr2D=np.array([[10,30,50],[20,40,60]])
#PRINTING ELEMENTS IN ARRAY
print("elements in arr1D=",arr1D)
print("elements in arr2D=",arr2D)
#INDEX
print("index:",arr1D[3])
#ROWS & COLUMS
print("row,col",arr2D[1:2])
#START:STOP
print("start:stop",arr1D[1:3])
#ROWS
print("row:",arr2D[0::])
#COLUMNS
print("column:",arr2D[::2])
```
